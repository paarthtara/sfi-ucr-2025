```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def number_of_final_states(k):
    upper_bound = int(np.ceil(k/2))
    sum = 0
    for s in range (1, upper_bound + 1):
        sum += math.comb(k - s + 1, s) * math.comb(k - 1, s - 1)
    return sum
```

```python
KSeries = np.arange(1, 20 + 1)
FinalStates = [number_of_final_states(k) for k in KSeries]
plt.plot(KSeries, FinalStates, marker='o')
```
